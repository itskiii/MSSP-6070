```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
word = input("Enter a word to search images for: ")
folder = word
os.makedirs(folder, exist_ok=True)
print(f"Images will be saved in: {os.path.abspath(folder)}")
```

```python
url = f"https://www.bing.com/images/search?q={word}"
headers = {"User-Agent": "Mozilla/5.0"}
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, "html.parser")
```

```python
count = 0
for img in soup.find_all("img"):
    src = img.get("src")
    if src and src.startswith("http"):
        try:
            img_data = requests.get(src, headers=headers, timeout=10).content
            filename = os.path.join(folder, f"{word}_{count+1}.jpg")
            with open(filename, "wb") as f:
                f.write(img_data)
            print(f"Saved {filename}")
            count += 1
            if count >= 5:
                break
        except:
            continue

if count > 0:
    print(f"\n✅ Download complete! {count} images saved in '{folder}'.")
else:
    print("\n⚠️ No images found. Try another keyword.")
```
