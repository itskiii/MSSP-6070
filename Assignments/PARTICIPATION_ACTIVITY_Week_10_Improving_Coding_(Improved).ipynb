```python
"""
Title: Student Test Scores Analysis
Description: This notebook calculates average scores for students
and visualizes the results in a bar chart.
"""

# --- Import required libraries ---
import pandas as pd
import matplotlib.pyplot as plt

# --- Create data ---
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Ella"],
    "Score1": [85, 90, 78, 92, 88],
    "Score2": [80, 85, 82, 95, 91]
}

df = pd.DataFrame(data)

# --- Validate data ---
if df.isnull().values.any():
    print("âš ï¸ Warning: Missing data detected!")
else:
    print("âœ… Data is complete.\n")

# --- Calculate average score ---
df["Average"] = df[["Score1", "Score2"]].mean(axis=1)

# --- Display results ---
print("Student Scores Summary:\n")
display(df)

# --- Plot average scores ---
plt.figure(figsize=(8, 5))
plt.bar(df["Name"], df["Average"], color="skyblue", edgecolor="black")
plt.title("Average Student Test Scores", fontsize=14, fontweight="bold")
plt.xlabel("Student Name")
plt.ylabel("Average Score")
plt.ylim(0, 100)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
